```python
df = (
    pd.read_csv("ocr-manual.csv")
    .loc[:, ["speed", "watts", "cadence", "time", "distance", "calories"]]
    .assign(velocity=lambda f: f.distance.diff().rolling(90).mean())
)
```

```python
p9.ggplot(df, p9.aes(x="time", y="speed")) + p9.geom_point()
p9.ggplot(df, p9.aes(x="time", y="watts")) + p9.geom_point()
p9.ggplot(df, p9.aes(x="time", y="distance")) + p9.geom_point()
p9.ggplot(df, p9.aes(x="time", y="velocity")) + p9.geom_point()
p9.ggplot(df, p9.aes(x="time", y="calories")) + p9.geom_point()
```
